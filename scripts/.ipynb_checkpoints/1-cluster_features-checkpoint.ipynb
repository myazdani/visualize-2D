```python
import os
from pylab import *
from sklearn import (manifold, datasets, decomposition, ensemble, lda,
                     random_projection)
from skimage.feature import hog
from skimage.filters import gaussian_filter
from skimage import color
from scipy import misc
from skimage.transform import resize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
import gc
```

```python
src_path = "../features/"
file_type = ".pkl"

feature_paths = []  
for root, dirs, files in os.walk(src_path):
    feature_paths.extend([os.path.join(root, f) for f in files if f.endswith(file_type)])
```

```python
hog_features_chunks = pd.read_pickle(feature_paths)
```

```python
hog_features_chunk_1 = pd.read_pickle("../features/hogs_chunk0.pkl")
hog_features_chunk_2 = pd.read_pickle("../features/hogs_chunk1.pkl")
hog_features_chunk_3 = pd.read_pickle("../features/hogs_chunk2.pkl")
```

```python
hog_features = pd.concat([hog_features_chunk_1, hog_features_chunk_2, hog_features_chunk_3])
del hog_features_chunk_1, hog_features_chunk_2, hog_features_chunk_3
gc.collect()
```

```python
image_features = array(list(hog_features["hogs"]))
image_features.shape
```

# Reduce dimensions and apply clustering

```python
pca = PCA(n_components=50)
X = pca.fit_transform(array(image_features))
```

```python
num_clusters = 100
kmeans = KMeans(n_clusters = num_clusters, n_init = 100, n_jobs=1)
#kmeans = KMeans(n_clusters = num_clusters)
clusters = kmeans.fit_predict(X)
```

```python
image_paths_hack = list(hog_features["image_paths"])
image_paths_rel = [image_path_hack.split("/")[-2] + "/" + image_path_hack.split("/")[-1] for image_path_hack in image_paths_hack]
image_paths_rel
```

```python
res_df = pd.DataFrame({'file_names' : image_paths_rel})
```

```python
res_df["clusters"] = clusters
```

```python
res_df.to_csv("../features/hogs_clusters.csv", header = True, index = False)
```

```python
image_paths_hack
```

# Apply 1D t-SNE

```python
tsne = manifold.TSNE(n_components=1, init='pca', random_state=0)
cluster_tsne = tsne.fit_transform(X)
```

```python
res_tsne = pd.DataFrame({"image_paths": image_paths_hack})
res_tsne["tsne.score"] = cluster_tsne
```

```python
res_tsne.to_csv("../features/hogs_tsne.csv", header = True, index = False)
```
